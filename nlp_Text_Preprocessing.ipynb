```python
def sent(text):
    root = []
    lemma = WordNetLemmatizer()
    x = word_tokenize(text)
    for i in x:
        if i not in set(stopwords.words("english")):
            token = lemma.lemmatize(i,pos="v")
            root.append(token)
    return root
            
            
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Sample text
#text = "Ram and .Shyam are, playing,. football"
def stems(text):
    # Split the text into words
    words = text.split()

    # Initialize WordNetLemmatizer
    lemmatizer = WordNetLemmatizer()

    # Define a set of English stopwords
    stop_words = set(stopwords.words("english"))

    # Process each word
    for word in words:
        # Remove punctuation and convert to lowercase
        word = word.lower()
        word = word.strip('.,')

        # Check if the word is not in the set of stopwords
        if word not in stop_words:
            # Lemmatize the word
            lemma = lemmatizer.lemmatize(word, pos="v")  # Use "v" for verbs
            print("{:<20} {:<20}".format(word, lemma))
```

```python
stems("Ram and shyam both are Best friends that are,,,,,,,,,,,................ going to school")
```
