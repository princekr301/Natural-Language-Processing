```python
def stemming(text):
    k=[]
    for i in text.split():
        if i not in stopwords.words("english"):
            k.append(i)
        else:
            k.append(" ")
    x = k[:]
    k.clear()
    k = " ".join(x).strip()
    k = re.sub(" +"," ",k)
    k = word_tokenize(k)
    st = [WordNetLemmatizer().lemmatize(i) for i in k]
    return st
```
